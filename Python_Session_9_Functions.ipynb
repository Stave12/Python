```python
def addition(a,b):
    print(a+b)
def subtraction(a,b):
    print(a-b)
def multiplication(a,b):
    print(a*b)
def division(a,b):
    print(a/b)
def calculator():
    print("option 1 is add")
    print("option 2 is sub")
    print("option 3 is mul")
    print("option 4 is duv")
    a=eval(input("enter a number:"))
    b=eval(input("enter a number:"))
    option=eval(input("enter option between 1 to 4"))
    if option == 1:
        addition(a,b)
    elif option == 2:
        subtraction(a,b)
    elif option == 3:
        multiplication(a,b)
    elif option ==4:
        division(a,b)
calculator()
        
    
```
