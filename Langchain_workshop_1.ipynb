Step-3:Text splitter

Chunking

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
# Chunk_Size
# Chunk_Overlap

splitter=CharacterTextSplitter(chunk_size=9000,chunk_overlap=500)
```

```python
docs=splitter.split_documents(documents)
len(docs)
```

```python
len(docs[0].page_content)
len(docs[1].page_content)

#len(docs[i].page_content) i=0 to 16
```
